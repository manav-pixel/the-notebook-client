```python
# Import necessary libraries
import matplotlib
import yaml
import pandas as pd
from pathlib import Path
from risk_validation.core.metrics.impl import pd as pd_metrics
from risk_validation.core.services.pd.metrics_service import PDMetricsService
from risk_validation.core.utils.plots import plot_gini, plot_roc, plot_ks_cdf_with_maxgap
from IPython.display import Image, display

# Set Matplotlib to non-interactive backend suitable for Jupyter
matplotlib.use("Agg")

# Load the dataset from the correct directory
DATA_DIR = Path('..') / 'data'
csv_path = DATA_DIR / 'sample.csv'
df = pd.read_csv(csv_path)

# Pre-filter the data to separate development and validation periods
dev_year = 2019
val_year = 2020
df_dev = df[df['score_year'] == dev_year].copy()
df_val = df[df['score_year'] == val_year].copy()

# Initialize metrics service with desired metrics
metric_ids = ['auc_roc', 'gini', 'ks']
service = PDMetricsService(metric_ids)

# Prepare parameters for metric calculation
params = {
    'y_col': 'default_flag',
    'p_col': 'score_pd',
    'score_col': 'score_pd'
}

# Compute metrics separately for development and validation datasets
dev_result = service.compute(df_dev, params)
val_result = service.compute(df_val, params)

# Print results
print("Development Period Metrics:", dev_result)
print("Validation Period Metrics:", val_result)

# Create directory for plots if not exists
plot_dir = Path('plots')
plot_dir.mkdir(exist_ok=True)

# Generate and display ROC curve plot
roc_plot_path = plot_roc(
    y_true=df_val['default_flag'],
    y_score=df_val['score_pd'],
    title='ROC Curve - Validation Data',
    out_path=plot_dir / 'roc.png'
)
display(Image(filename=str(roc_plot_path)))

# Generate and display Gini plot for validation data
gini_plot_path = plot_gini(
    y_true=df_val['default_flag'],
    y_score=df_val['score_pd'],
    title='Gini Curve - Validation Data',
    out_path=plot_dir / 'gini.png'
)
display(Image(filename=str(gini_plot_path)))

# Generate and display KS CDF plot for validation data
ks_plot_path, _ = plot_ks_cdf_with_maxgap(
    y_true=df_val['default_flag'],
    y_score=df_val['score_pd'],
    title='KS CDF - Validation Data',
    out_path=plot_dir / 'ks_cdf.png'
)
display(Image(filename=str(ks_plot_path)))
```
